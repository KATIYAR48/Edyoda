```python
import random

class FoodItem:
    def __init__(self, food_id, name, quantity, price, discount, stock):
        self.food_id = food_id
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock

class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        self.orders = []

    def place_order(self, items):
        order = Order(items)
        self.orders.append(order)
        return order

    def update_profile(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password

class Order:
    def __init__(self, items):
        self.items = items

    def calculate_total_payable(self):
        total_payable = 0
        for item in self.items:
            item_price = item.price * (1 - item.discount / 100)
            total_payable += item_price
        return total_payable

class Restaurant:
    def __init__(self):
        self.food_menu = []

    def add_food_item(self, name, quantity, price, discount, stock):
        food_id = self.generate_food_id()
        food_item = FoodItem(food_id, name, quantity, price, discount, stock)
        self.food_menu.append(food_item)
        return food_item

    def edit_food_item(self, food_id, name, quantity, price, discount, stock):
        for food_item in self.food_menu:
            if food_item.food_id == food_id:
                food_item.name = name
                food_item.quantity = quantity
                food_item.price = price
                food_item.discount = discount
                food_item.stock = stock
                return food_item
        return None

    def remove_food_item(self, food_id):
        for food_item in self.food_menu:
            if food_item.food_id == food_id:
                self.food_menu.remove(food_item)
                return True
        return False

    def get_food_menu(self):
        return self.food_menu

    def generate_food_id(self):
        food_id = random.randint(1, 100)
        while self.check_food_id_exists(food_id):
            food_id = random.randint(1, 100)
        return food_id

    def check_food_id_exists(self, food_id):
        for food_item in self.food_menu:
            if food_item.food_id == food_id:
                return True
        return False

# Function to print the food menu
def print_food_menu(food_menu):
    print("Food Menu:")
    for food_item in food_menu:
        print(f"{food_item.food_id}. {food_item.name} ({food_item.quantity}) [INR {food_item.price}], Discount: {food_item.discount}, Stock: {food_item.stock}")

# Function to print the order history
def print_order_history(order_history):
    print("Order History:")
    for order in order_history:
        print("Order:")
        for item in order.items:
            print(f"ID: {item.food_id}, Name: {item.name}, Quantity: {item.quantity}, Price: {item.price}, Discount: {item.discount}")
        print("Total Payable: INR", order.calculate_total_payable())
        print("--------------")

# Admin credentials
admin_credentials = {
    "username": "admin",
    "password": "admin123"
}

restaurant = Restaurant()
user = None
order_history = []

# Default food items
restaurant.add_food_item("Tandoori Chicken", "4 pieces", 240, 0, 100)
restaurant.add_food_item("Vegan Burger", "1 Piece", 320, 0, 50)
restaurant.add_food_item("Truffle Cake", "500gm", 900, 0, 20)

# User registration count
user_registration_count = 0

while True:
    print("Welcome to the Food Ordering App!")
    print("1. Admin Login")
    print("2. User Login")
    print("3. User Registration")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        print("Admin Login")
        username = input("Enter your username: ")
        password = input("Enter your password: ")

        if username == admin_credentials["username"] and password == admin_credentials["password"]:
            print("Admin login successful!")

            while True:
                print("Admin Menu:")
                print("1. Add Food Item")
                print("2. Edit Food Item")
                print("3. View Food Menu")
                print("4. Remove Food Item")
                print("5. Log Out")

                admin_choice = input("Enter your choice: ")

                if admin_choice == "1":
                    name = input("Enter the food item name: ")
                    quantity = input("Enter the quantity: ")
                    price = float(input("Enter the price: "))
                    discount = float(input("Enter the discount: "))
                    stock = int(input("Enter the stock amount: "))

                    food_item = restaurant.add_food_item(name, quantity, price, discount, stock)
                    print(f"Food item added! ID: {food_item.food_id}")

                elif admin_choice == "2":
                    food_id = int(input("Enter the food item ID: "))
                    name = input("Enter the new name: ")
                    quantity = input("Enter the new quantity: ")
                    price = float(input("Enter the new price: "))
                    discount = float(input("Enter the new discount: "))
                    stock = int(input("Enter the new stock amount: "))

                    food_item = restaurant.edit_food_item(food_id, name, quantity, price, discount, stock)
                    if food_item:
                        print("Food item edited successfully!")
                    else:
                        print("Food item not found!")

                elif admin_choice == "3":
                    food_menu = restaurant.get_food_menu()
                    print_food_menu(food_menu)

                elif admin_choice == "4":
                    food_id = int(input("Enter the food item ID: "))
                    result = restaurant.remove_food_item(food_id)
                    if result:
                        print("Food item removed successfully!")
                    else:
                        print("Food item not found!")

                elif admin_choice == "5":
                    print("Admin logged out!")
                    break

                else:
                    print("Invalid choice!")

        else:
            print("Invalid username or password!")

    elif choice == "2":
        print("User Login")
        if user is None:
            print("Please register first!")
            continue

        while True:
            print("1. Place New Order")
            print("2. Order History")
            print("3. Update Profile")
            print("4. Log Out")

            user_choice = input("Enter your choice: ")

            if user_choice == "1":
                food_menu = restaurant.get_food_menu()
                print_food_menu(food_menu)

                selected_items = []
                while True:
                    order_input = input("Select the food items by entering their IDs (comma-separated): ")
                    item_ids = order_input.split(",")
                    for item_id in item_ids:
                        for food_item in food_menu:
                            if food_item.food_id == int(item_id.strip()):
                                selected_items.append(food_item)
                    more_items = input("Do you want to add more items? (y/n): ")
                    if more_items.lower() != "y":
                        break

                if selected_items:
                    print("Selected Items:")
                    print_food_menu(selected_items)

                    place_order = input("Do you want to place this order? (y/n): ")
                    if place_order.lower() == "y":
                        order = user.place_order(selected_items)
                        order_history.append(order)
                        print("Order placed successfully!")
                        print("Total Payable: INR", order.calculate_total_payable())
                        print("--------------")
                else:
                    print("No items selected!")

            elif user_choice == "2":
                print_order_history(order_history)

            elif user_choice == "3":
                print("Update Profile")
                new_full_name = input("Enter your new full name: ")
                new_phone_number = input("Enter your new phone number: ")
                new_email = input("Enter your new email: ")
                new_address = input("Enter your new address: ")
                new_password = input("Enter your new password: ")

                user.update_profile(new_full_name, new_phone_number, new_email, new_address, new_password)
                print("Profile updated successfully!")

            elif user_choice == "4":
                print("Logged out!")
                break

            else:
                print("Invalid choice!")

    elif choice == "3":
        if user_registration_count < 5:
            print("User Registration")
            full_name = input("Enter your full name: ")
            phone_number = input("Enter your phone number: ")
            email = input("Enter your email: ")
            address = input("Enter your address: ")
            password = input("Enter your password: ")

            user = User(full_name, phone_number, email, address, password)
            user_registration_count += 1
            print("User registration successful!")

        else:
            print("User registration limit reached!")

    elif choice == "4":
        print("Thank you for using the Food Ordering App!")
        break

    else:
        print("Invalid choice!")
r
```
